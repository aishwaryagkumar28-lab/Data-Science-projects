# Placement

```python
import pandas as pd
```

```python
# Load the dataset
file_path = 'https://raw.githubusercontent.com/YBIFoundation/Dataset/main/Placement.csv'  # Update this path to where your file is located
data = pd.read_csv(file_path)
```

```python
# Display the first few rows of the dataset
print(data.head())
```

```python
print(data.info())
```

```python
x = data['Placement']
y = data['IQ']
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
# Reshape x_train and x_test to 2D arrays
x_train_reshaped = x_train.values.reshape(-1, 1)  # -1 infers the number of samples
x_test_reshaped = x_test.values.reshape(-1, 1)

x_train = scaler.fit_transform(x_train_reshaped)
x_test = scaler.transform(x_test_reshaped)
```

```python
from sklearn.linear_model import LogisticRegression

# Instantiate the model
model = LogisticRegression(random_state=42)

# Train the model
model.fit(x_train, y_train)
```

```python
from sklearn.metrics import accuracy_score, classification_report

# Predict on the test set
y_pred = model.predict(x_test)
```

```python
# Calculate accuracy
accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy}')
```

```python
# Print classification report
print(classification_report(y_test, y_pred))
```
